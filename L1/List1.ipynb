```julia
epsil(a) = begin
    n=0
    while a+2. ^(-n)>a
        n+=1
    end
    n-1
end
```

```julia
epsil(1)
```

```julia
epsil.([10,100,1000,10^4])
```

```julia
rek1p(N;x=1,x=1/5,a=5,b=26,c=5)=begin
    N=N-1
    if N==0
        return [x,x]
    end
    rek1p(N,x=x, x=(b*x-a*x)/c,a=a,b=b,c=c)
end
```

```julia
tmp=rek1p(30)
```

```julia
rek1t(N;x=1,x=1/5,a=5,b=26,c=5)=begin
    N=N-1
    if N==0
        return [x,x]
    end
    rek1t(N,x=(b*x-c*x)/a, x=x, a=a,b=b,c=c)
end
```

```julia
rek1t(30,x=tmp[1],x=tmp[2])
```

```julia
g.(tmp)
```

```julia
g.(tmp)
```

```julia
plot(tmp,g, lab="g1")
plot!(tmp,g, lab="g2")
```
