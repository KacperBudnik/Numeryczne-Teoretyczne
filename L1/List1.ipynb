```julia
epsil(a) = begin
    n=0
    while a+2. ^(-n)>a
        n+=1
    end
    n-1
end
```

```julia
epsil(1)
```

```julia
epsil.([10,100,1000,10^4])
```

```julia
n = rand(UInt8)
Xₙ₋₁ = (1//BigInt(5))^(n-1)
Xₙ   = (1//BigInt(5))^n
Xₙ₊₁ = (1//BigInt(5))^(n+1);
```

```julia
5Xₙ₋₁-26Xₙ+5Xₙ₊₁==0
```

```julia
rek1p(N;x₀=1,x₁=1/5,a=5,b=26,c=5)=begin
    N=N-1
    if N==0
        return [x₀,x₁]
    end
    rek1p(N,x₀=x₁, x₁=(b*x₁-a*x₀)/c,a=a,b=b,c=c)
end
```

```julia
tmp=rek1p(30)
```

```julia
rek1t(N;x₀=1,x₁=1/5,a=5,b=26,c=5)=begin
    N=N-1
    if N==0
        return [x₀,x₁]
    end
    rek1t(N,x₀=(b*x₀-c*x₁)/a, x₁=x₀, a=a,b=b,c=c)
end
```

```julia
rek1t(30,x₀=tmp[1],x₁=tmp[2])
```

# Zad 4

```julia
g₁(x)=(x-1)^4
g₂(x)=x^4-4x^3+6x^2-4x+1;
```

```julia
tmp=1-10^-3:10^-4:1+10^-3
```

```julia
g₁.(tmp)
```

```julia
g₂.(tmp)
```

```julia
plot(tmp,g₁, lab="g1")
plot!(tmp,g₂, lab="g2")
```

```julia
plot(tmp,x->g₁(x)+1e-16, lab="g1")
plot!(tmp,x->g₂(x)+1e-16, lab="g2")
plot!(yscale=:log)
```
