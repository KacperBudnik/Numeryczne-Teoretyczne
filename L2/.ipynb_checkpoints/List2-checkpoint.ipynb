```julia
insert_and_dedup!(v::Vector, x) = (splice!(v, searchsorted(v,x), [x]); v)
```

```julia
method(;N=N,f=f)=begin
    X=[f(i) for i in N:-1:1]
    while length(X)>1
        tmp=popfirst!(X)+popfirst!(X)
        insert_and_dedup!(X,tmp)
    end
    X[1]
end
```

```julia
using DataStructures
```

```julia
methodv2(;N=N,f=f)=begin
    X=SortedSet([f(i) for i in N:-1:1])
    while length(X)>1
        tmp=pop!(X)+pop!(X)
        push!(X,tmp)
    end
    pop!(X)
end
```

# Zad 3

```julia
bisek(;f,a,b,=NaN,=NaN)=begin
    if !xor(===NaN, ===NaN)
        error("Algorytm potrzebuje dokadnie jednej wartoci ze zbioru {, }!")
    end
    fa=f(a)
    fb=f(b)
    if fa*fb >0
        error("Funkcja na przegach przyjmuj ten sam znak!")
    end

    c=(b+a)/2
    fc=f(c)
    if ===NaN
        while abs(fc)>
            if fa*fc <0
                b=c
            else
                a=c
            end
            c=(b+a)/2
            fc=f(c)
        end
    else
        while b-a>2*
            if fa*fc <0
                b=c
            else
                a=c
            end
            c=(b+a)/2
            fc=f(c)
        end
    end
    (c,f(c))
end
```

```julia
bisek(f=sin, a=-1, b=3/2, =1e-4)
```

```julia
bisek(f=sin, a=-1, b=3/2, =1e-6)
```

# Zad 4

```julia
newton(;f,f,x,)=begin
    x=[x]
    fx=[f(x)]
    while abs(fx[end])>
        push!(x,x[end]-fx[end]/f(x[end]))
        push!(fx,f(x[end]))
    end
    [x[end],hcat(x,fx)]
end
```

```julia
newton(f=x->sin(x), f=x->-cos(x),x=0.1,=0.01)[1]
```

# Zad 5

```julia
f_zerowe(x)=log(x)-x+4
```

```julia
plot(0.001:0.001:6,f_zerowe, leg=false)
```

Poniewa
$$ \lim\limits_{x\to\infty}\ln(x)-x+4=-\infty=\lim\limits_{x\to0}\ln(x)-x+4$$
oraz $\ln(1)-1+4=3>0$, to s przynajmniej 2 miejsca zerowe w prziedziale $(0,1)$ i $(1,\infty)$. W pierwszym przedziale $x$ jest may, wic miejsce zerowe bdzie w przyblizeniu w punkcie $x_1$, gdzie $\ln(x_1)=-4$, czyli $x_1\approx e^{-4}$.

Drugie miejsce zerowe:

$\ln(4)\approx 1.5$

$\ln(x)^\prime(4)=1/4$

$\frac{3}{4} z = 1.5 \iff z =2$

$x=4+2=6$

```julia
f_zerowe(exp(-4)/2)
```

```julia
f_zerowe(exp(-4)*2)
```
