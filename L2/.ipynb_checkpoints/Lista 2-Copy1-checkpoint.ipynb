```julia
method₁(;N=N,f=f)=begin
    sum(f(i) for i in 1:N)
end
```

```julia
method₂(;N=N,f=f)=begin
    tmp=0
    for i in 1:N
        tmp+=f(i)
    end
    tmp
end
```

```julia
method₃(;N=N,f=f)=begin
    tmp=0
    for i in N:-1:1
        tmp+=f(i)
    end
    tmp
end
```

```julia
@time π^2/6-method₂()
```

```julia
@time π^2/6-method₃()
```

```julia
insert_and_dedup!(v::Vector, x) = (splice!(v, searchsorted(v,x), [x]); v)
```

```julia
?searchsorted
```

```julia
method₄(;N=N,f=f)=begin
    X=[f(i) for i in N:-1:1]
    while length(X)>1
        tmp=popfirst!(X)+popfirst!(X)
        insert_and_dedup!(X,tmp)
    end
    X[1]
end
```

```julia
@time method₄(N=2^10)
```
