```julia
euler_method(;f,a,b,h, y₀, end_point=:false)=begin
    X=a:h:b
    Y=zeros(length(X))
    Y[1]=y₀
    for i in 1:length(X)-1
        Y[i+1]=Y[i]+h*f(X[i],Y[i])
    end
    if end_point && X[end] != b
        h=(b-X[end])
        push!(Y, Y[end]+h*f( X[end],Y[end]))
        X=union(X,b)
    end
    hcat(X,Y)
end
```

```julia
heun_method(;f,a,b,h, y₀, end_point=:false)=begin
    X=a:h:b
    Y=zeros(length(X))
    Y[1]=y₀
    for i in 1:length(X)-1
        Y[i+1]=Y[i]+h*f( X[i]+h/2 , Y[i]+h/2*f( X[i],Y[i]))
    end
    if end_point && X[end] != b
        h=(b-X[end])
        push!(Y, Y[end]+h*f( X[end]+h/2 , Y[end]+h/2*f( X[end],Y[end])) )
        X=union(X,X[end]+h)
    end
    hcat(X,Y)
end
```

```julia
RK_method(;f,a,b,h, y₀, end_point=:false)=begin
    X=a:h:b
    Y=zeros(length(X))
    Y[1]=y₀
    for i in 1:length(X)-1
        k₁=h*f(X[i]    , Y[i]     )
        k₂=h*f(X[i]+h/2, Y[i]+k₁/2)
        k₃=h*f(X[i]+h/2, Y[i]+k₂/2)
        k₄=h*f(X[i]+h  , Y[i]+k₃  )
        Y[i+1]=Y[i]+(k₁+2k₂+2k₃+k₄)/6
    end
    if end_point && X[end] != b
        h=(b-X[end])
        k₁=h*f(X[end]    , Y[end]     )
        k₂=h*f(X[end]+h/2, Y[end]+k₁/2)
        k₃=h*f(X[end]+h/2, Y[end]+k₂/2)
        k₄=h*f(X[end]+h  , Y[end]+k₃  )
        push!(Y,Y[end]+(k₁+2k₂+2k₃+k₄)/6 )
        X=union(X,b)
    end
    hcat(X,Y)
end
```

```julia
Richardson(;f,h=0.01, method=:Euler, a=0, b=1, y₀)=begin
    order=1
    if method==:Euler
        order=1
        A=euler_method(f=f, y₀=y₀, a=a, b=b, h=h)
        B=euler_method(f=f, y₀=y₀, a=a, b=b, h=h/2)
    elseif method==:Heun
        order=2
        A=heun_method(f=f, y₀=y₀, a=a, b=b, h=h)
        B=heun_method(f=f, y₀=y₀, a=a, b=b, h=h/2)
    elseif method==:RK
        order=4
        A=RK_method(f=f, y₀=y₀, a=a, b=b, h=h)
        B=RK_method(f=f, y₀=y₀, a=a, b=b, h=h/2)
    else
        error("Błędna metoda")
    end
    
    (2^order*B[1:2:end,2]-A[:,2])/(2^order-1)
    
end
```

```julia
euler_method(;f,a,b,h, y₀)=begin
    X=a:h:b
    Y=zeros(length(X), length(y₀))
    Y[1,:]=y₀
    for i in 1:length(X)-1
        Y[i+1,:] = Y[i, :] + h.* f(X[i],Y[i,:])
    end
    Y    
end
```

```julia
heun_method(;f,a,b,h, y₀)=begin
    X=a:h:b
    Y=zeros(length(X), length(y₀))
    Y[1,:]=y₀
    for i in 1:length(X)-1
        Y[i+1, :]=Y[i,:]+h.*f( X[i]+h/2 , Y[i,:]+ (h/2) .*f(X[i],Y[i]))
    end
    Y
end
```

```julia
RK_method(;f,a,b,h, y₀)=begin
    X=a:h:b
    Y=zeros(length(X), length(y₀))
    Y[1,:]=y₀
    for i in 1:length(X)-1
        k₁=h.*f(X[i]    , Y[i,:]     )
        k₂=h.*f(X[i]+h/2, Y[i,:]+k₁/2)
        k₃=h.*f(X[i]+h/2, Y[i,:]+k₂/2)
        k₄=h.*f(X[i]+h  , Y[i,:]+k₃  )
        Y[i+1,:]=Y[i,:]+(k₁+2 .*k₂+2 .*k₃+k₄) ./6
    end
    Y
end
```
